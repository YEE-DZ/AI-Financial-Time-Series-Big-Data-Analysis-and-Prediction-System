# 상대적 수익률 (코스피지수에 따른 수익률계산)

* 상대적 수익률=(
벤치마크의 수익률
포트폴리오 또는 종목의 수익률−벤치마크의 수익률
​
 )×100%

예를 들어, 특정 기간 동안 포트폴리오의 수익률이 15%이고, 동일 기간 동안 코스피 지수의 수익률이 10%라면, 상대적인 수익률은 다음과 같이 계산됩니다.

상대적 수익률
=
(
15
−
10
10
)
×
100
%
=
50
%

```python
#### 상대적 수익률(코스피,코스닥과 비교한 수익률) 함수
def calculate_relative_return(portfolio_return, benchmark_return):
    relative_return = ((portfolio_return - benchmark_return) / benchmark_return) * 100
    return relative_return

# 예시: 포트폴리오 수익률과 코스피 지수 수익률
portfolio_return = 15  # 포트폴리오 수익률
benchmark_return = 10  # 코스피 지수 수익률

# 상대적인 수익률 계산
relative_return = calculate_relative_return(portfolio_return, benchmark_return)
print(f"상대적인 수익률: {relative_return:.2f}%")
```

```python
## test 날짜기간 동안의 코스피 지수 비교해봐야함.
from datetime import datetime

# 특정 기간의 시작일과 종료일 설정 (예시: '2023-01-01'부터 '2023-12-31')
start_date = '2023-01-01'
end_date = '2023-12-31'

# 날짜 기간 계산
datetime_format = "%Y-%m-%d"
start_datetime = datetime.strptime(start_date, datetime_format)
end_datetime = datetime.strptime(end_date, datetime_format)
days_diff = (end_datetime - start_datetime).days


# 특정 기간 동안의 코스피 지수 초기 값과 최종 값 가져오기
kospi_period = kospi[start_date:end_date] ##코스피
initial_kospi = kospi_period.iloc[0]['Close'] ###########아마 시가총액으로 해야할듯. 'MarCap'!!!!!!!!!!!!!!!!!
final_kospi = kospi_period.iloc[-1]['Close']

kosdaq_period = kosdaq[start_date:end_date] #코스닥
initial_kosdaq = kosdaq_period.iloc[0]['Close']
final_kosdaq = kosdaq_period.iloc[-1]['Close']

# 수익률 계산
kospi_return = ((final_kospi - initial_kospi) / initial_kospi) * 100  #이게 위의 benchmark_return
kosdaq_return = ((final_kosdaq - initial_kosdaq) / initial_kosdaq) * 100  #이게 위의 benchmark_return

print(f"특정 기간({days_diff}일) 동안의 <<코스피>> 지수 수익률: {kospi_return:.2f}%")
print(f"특정 기간({days_diff}일) 동안의 <<코스닥>> 지수 수익률: {kosdaq_return:.2f}%")
```

## 19종목 거래기간 날짜 수집 및 시장분류(코스피,코스닥)

```python
import pandas as pd
from os.path import join

def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

file_directories = {
    "재은": "/content/drive/MyDrive/0_Capstone/data/model_signal/재은/",
    "효림": "/content/drive/MyDrive/0_Capstone/data/model_signal/효림/",
    "희선": "/content/drive/MyDrive/0_Capstone/data/model_signal/희선/"
}

file_names = [
    "new_KR7005930003.csv",
    "new_KR7000250001.csv",
    "new_KR7036570000.csv",
    "new_KR7051910008.csv",
    "new_KR7066700006.csv",
    "new_KR7066970005.csv",
    "new_KR7068760008.csv",
    "new_KR7078600004.csv",
    "new_KR7096770003.csv",
    "new_KR7128940004.csv",
    "new_KR7185750007.csv",
    "new_KR7192080000.csv",
    "new_KR7207940008.csv",
    "new_KR7225570001.csv",
    "new_KR7247540008.csv",
    "new_KR7251270005.csv",
    "new_KR7263750002.csv",
    "new_KR7293490009.csv",
    "new_KR7373220003.csv",
]

for person, directory in file_directories.items():
    person_df = pd.DataFrame()
    for file_name in file_names:
        file_path = join(directory, file_name)
        df = load_data(file_path)

        # Extracting first and last dates
        first_date = df['Timestamp'].min()
        last_date = df['Timestamp'].max()

        # Get the stock name from file_name (assuming format remains consistent)
        stock_name = file_name.split('_')[1].split('.')[0]

        # Create a new DataFrame to store first and last dates along with stock name
        date_df = pd.DataFrame({'Stock': [stock_name], 'First Date': [first_date], 'Last Date': [last_date]})

        # Concatenate date_df to the respective person_df
        person_df = pd.concat([person_df, date_df])

    # Reset index for the person_df
    person_df.reset_index(drop=True, inplace=True)

    # Save person_df separately as ESN_date, RNN_date, and Cls_date
    if person == "재은":
        ESN_date = person_df.copy()
    elif person == "효림":
        RNN_date = person_df.copy()
    elif person == "희선":
        Cls_date = person_df.copy()

# print("ESN_date:")
# print(ESN_date)
# print("\nRNN_date:")
# print(RNN_date)
# print("\nCls_date:")
# print(Cls_date)
```

```python
# Creating a list of KOSPI stocks
kospi_stocks = [
    'KR7096770003', 'KR7051910008', 'KR7373220003', 'KR7128940004',
    'KR7207940008', 'KR7185750007', 'KR7192080000', 'KR7251270005',
    'KR7036570000', 'KR7005930003'
]

# Iterate through each _date dataframe to add the 'stock_type' column
for df_name in [ESN_date, RNN_date, Cls_date]:
    df_name['stock_type'] = df_name['Stock'].apply(lambda x: 'kospi' if x in kospi_stocks else 'kosdaq')

# # Displaying the updated dataframes
# print("ESN_date:")
# print(ESN_date)
# print("\nRNN_date:")
# print(RNN_date)
# print("\nCls_date:")
# print(Cls_date)
```

## 종목 거래 기간 별 시장 수익률(변동률)

```python
from datetime import datetime

# 함수를 이용하여 날짜 간격 계산
def calculate_days(start_date, end_date):
    datetime_format = "%Y-%m-%d"
    start_datetime = datetime.strptime(start_date, datetime_format)
    end_datetime = datetime.strptime(end_date, datetime_format)
    return (end_datetime - start_datetime).days

# 각 _date 프레임을 순회하면서 처리
for df_name in [ESN_date, RNN_date, Cls_date]:
    for stock, group in df_name.groupby('Stock'):
        start_date = group['First Date'].values[0]
        end_date = group['Last Date'].values[0]

        # start_date와 end_date로부터 days_diff를 계산하여 trade_days 컬럼에 추가
        # df_name.loc[df_name['Stock'] == stock, 'start_date'] = start_date
        # df_name.loc[df_name['Stock'] == stock, 'end_date'] = end_date
        df_name.loc[df_name['Stock'] == stock, 'trade_days'] = calculate_days(start_date, end_date)

        # stock_type에 따라 kospi_return 또는 kosdaq_return을 market_return 컬럼에 할당
        stock_type = group['stock_type'].values[0]
        if stock_type == 'kospi':
            kospi_period = kospi[start_date:end_date] ##코스피
            initial_kospi = kospi_period.iloc[0]['MarCap'] #시가총액으로 계산
            final_kospi = kospi_period.iloc[-1]['MarCap']
            kospi_return = ((final_kospi - initial_kospi) / initial_kospi) * 100  #이게 위의 benchmark_return
            market_return = kospi_return
        else:
            kosdaq_period = kosdaq[start_date:end_date] #코스닥
            initial_kosdaq = kosdaq_period.iloc[0]['MarCap']
            final_kosdaq = kosdaq_period.iloc[-1]['MarCap']
            kosdaq_return = ((final_kosdaq - initial_kosdaq) / initial_kosdaq) * 100  #이게 위의 benchmark_return
            market_return = kosdaq_return

        # market_return 값을 'market_return' 컬럼에 추가
        df_name.loc[df_name['Stock'] == stock, 'market_return'] = market_return

# 수정된 _date 프레임 출력
print("ESN_date:")
print(ESN_date)
print("\nRNN_date:")
print(RNN_date)
print("\nCls_date:")
print(Cls_date)
```

## 종목별 상대적 수익률 계산

```python
#### 상대적 수익률(코스피,코스닥과 비교한 수익률) 함수
def calculate_relative_return(portfolio_return, benchmark_return):
    relative_return = ((portfolio_return - benchmark_return) / benchmark_return) * 100
    return relative_return

for df_name in [ESN_date1, RNN_date1, Cls_date1]:
    for stock, group in df_name.groupby('Stock'):
        portfolio_return = group['stock_return'].values[0]
        benchmark_return = group['market_return'].values[0]
        relative_return = calculate_relative_return(portfolio_return, benchmark_return)
        df_name.loc[df_name['Stock'] == stock, 'relative_return'] = relative_return
        original_return = group['stock_return'].values[0]
        df_name.loc[df_name['Stock'] == stock, 'diff_return'] = -1*(original_return - relative_return) ## 기존 수익 대비 상대적 수익률 추이

display(ESN_date1)
display(RNN_date1)
display(Cls_date1)
```

```python
import matplotlib.pyplot as plt

# 데이터프레임 목록과 이름 목록 생성
dataframes = [ESN_date1, RNN_date1, Cls_date1]
names = ['ESN_date1', 'RNN_date1', 'Cls_date1']

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, (df, name) in enumerate(zip(dataframes, names)):
    # 최대, 최소값 찾기
    max_val = df['relative_return'].max()
    min_val = df['relative_return'].min()
    diff_val = max_val - min_val

    # 주어진 값에 해당하는 종목 코드 찾기
    stock_with_max = df.loc[df['relative_return'] == max_val, 'Stock'].values
    stock_with_min = df.loc[df['relative_return'] == min_val, 'Stock'].values

    # 막대 그래프 그리기
    axs[i].bar('MAX', max_val, color='blue')
    axs[i].bar('MIN', min_val, color='red')

    # 상대 수익률 값과 종목 이름 추가
    axs[i].text('MAX', max_val, f"MAX: {max_val:.2f}\n{stock_with_max[0]}", ha='center', va='bottom', color='black', fontsize=10)
    axs[i].text('MIN', min_val, f"MIN: {min_val:.2f}\n{stock_with_min[0]}", ha='center', va='bottom', color='black', fontsize=10)

    axs[i].set_title(f'Relative Return for {name}')
    axs[i].set_ylabel('Relative Return')

    # 점선 추가를 위해 y 축 범위 조정
    axs[i].set_ylim(min_val - 0.5 * diff_val, max_val + 0.5 * diff_val)
    axs[i].axhline(y=max_val, color='blue', linestyle='--', linewidth=1)  # 최대값 점선
    axs[i].axhline(y=min_val, color='red', linestyle='--', linewidth=1)   # 최소값 점선

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터프레임 목록과 이름 목록 생성
dataframes = [ESN_date1, RNN_date1, Cls_date1]
names = ['ESN_date1', 'RNN_date1', 'Cls_date1']

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

all_vals = []  # 최대, 최소값을 저장할 리스트

for i, (df, name) in enumerate(zip(dataframes, names)):
    # 최대, 최소값 찾기
    max_val = df['relative_return'].max()
    min_val = df['relative_return'].min()
    diff_val = max_val - min_val

    # 값 리스트에 추가
    all_vals.extend([max_val, min_val])

# 전체 데이터의 최소값과 최대값
global_max = max(all_vals)
global_min = min(all_vals)

for i, (df, name) in enumerate(zip(dataframes, names)):
    # 최대, 최소값 찾기
    max_val = df['relative_return'].max()
    min_val = df['relative_return'].min()
    diff_val = max_val - min_val

    # 막대 그래프 그리기
    axs[i].bar('MAX', max_val, color='blue')
    axs[i].bar('MIN', min_val, color='red')

    # 상대 수익률 값과 종목 이름 추가
    axs[i].text('MAX', max_val, f"MAX: {max_val:.2f}\n{stock_with_max[0]}", ha='center', va='bottom', color='black', fontsize=10)
    axs[i].text('MIN', min_val, f"MIN: {min_val:.2f}\n{stock_with_min[0]}", ha='center', va='bottom', color='black', fontsize=10)

    axs[i].set_title(f'Relative Return for {name}')
    axs[i].set_ylabel('Relative Return')

    # 점선 추가를 위해 y 축 범위 조정
    axs[i].set_ylim(global_min-100000, global_max+100000)
    axs[i].axhline(y=max_val, color='blue', linestyle='--', linewidth=1)  # 최대값 점선
    axs[i].axhline(y=min_val, color='red', linestyle='--', linewidth=1)   # 최소값 점선

plt.tight_layout()
plt.show()
```

### 종목, 상대 수익률 비교 시각화

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

for i, (df_name, title) in enumerate([(ESN_date1, 'ESN_date1'), (RNN_date1, 'RNN_date1'), (Cls_date1, 'Cls_date1')]):
    stock_return = []
    relative_return = []
    stocks = []

    for stock, group in df_name.groupby('Stock'):
        stocks.append(stock)
        stock_return.append(group['stock_return'].values[0])
        relative_return.append(group['relative_return'].values[0])

    axs[0, i].bar(stocks, stock_return, label='Stock Return')
    axs[0, i].bar(stocks, relative_return, label='Relative Return')
    axs[0, i].set_title(f'{title} Stock vs Relative Return')
    axs[0, i].set_xlabel('Stock')
    axs[0, i].set_ylabel('Return')
    axs[0, i].legend()
    axs[0, i].tick_params(axis='x', rotation=90)

    axs[1, i].bar(stocks, relative_return, label='Relative Return')
    axs[1, i].set_title(f'{title} Relative Return')
    axs[1, i].set_xlabel('Stock')
    axs[1, i].set_ylabel('Return')
    axs[1, i].legend()
    axs[1, i].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
